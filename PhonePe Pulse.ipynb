```python
root_dir = r'C:\clone_files\pulse\data'                 #C:\clone_files\pulse\data\aggregated\transaction\country\india\state

#Initialize empty list to hold dictionaries of data for each JSON file
data1_list = []

#Loop over all the state folders
state_dir = os.path.join(root_dir, 'aggregated', 'transaction', 'country', 'india', 'state')
for state in os.listdir(state_dir):
    state_path = os.path.join(state_dir, state)
    if os.path.isdir(state_path):
        
        #Loop over all the year folders
        for year in os.listdir(state_path):
            year_path = os.path.join(state_path, year)
            if os.path.isdir(year_path):
                
                #Loop over all the json files (one for each quarter)
                for json_file in os.listdir(year_path):
                    if json_file.endswith('.json'):
                        with open(os.path.join(year_path, json_file)) as f:
                            data = json.load(f)
                            
                            #Extract the data we're interested in
                            for transaction_data in data['data']['transactionData']:
                                row_dict = {
                                    'States' : state,
                                    'Transaction_year' : year,
                                    'Quarters' : int(json_file.split('.')[0]),
                                    'Transaction_Type' : transaction_data['name'],
                                    'Transaction_Count' : transaction_data['paymentInstruments'][0]['count'],
                                    'Transaction_Amount' : transaction_data['paymentInstruments'][0]['amount']
                                    
                                }
                                data1_list.append(row_dict)
                                
# Covert list of dictionaries to DataFrame
df1 = pd.DataFrame(data1_list)

# Convert DataFrame into CSV
df1.to_csv('data_agg_trans.csv', index=False)
```

```python
root_dir = r'C:\clone_files\pulse\data'

state_path =  os.path.join(root_dir, 'aggregated', 'user', 'country', 'india', 'state')
state_list =  os.listdir(state_path)
data2_list = {
              'state': [], 'Year': [],  'Quarter': [], 
              'Brand': [], 'Transaction_count': [], 'Percentage': []
             }

for state in state_list: 
    year_path = state_path + '/'+ state + '/' 
    year_list = os.listdir (year_path)
    
    for year in year_list:
        quarter_path = year_path + year + '/' 
        quarter_list = os.listdir (quarter_path)
        
        for quarter in quarter_list: 
            json_path = quarter_path + quarter 
            df = pd.read_json(json_path)
            
            try:
                for user_data in df['data']['usersByDevice']:
                    brand = user_data['brand'] 
                    count = user_data['count']
                    percent = user_data['percentage']
                    
                    data2_list['state'].append(state)
                    data2_list['Year'].append(year)
                    data2_list['Quarter'].append(int(quarter.removesuffix('.json')))
                    data2_list['Brand'].append(brand)
                    data2_list['Transaction_count'].append(count)
                    data2_list['Percentage'].append(percent)
            except:
                pass

df2 = pd.DataFrame(data2_list)
df2.to_csv('data_agg_user.csv', index=False)
```

```python
root_dir = r'C:\clone_files\pulse\data'                     #C:\clone_files\pulse\data\map\user\hover\country\india\state

data4_list = []

state_dir = os.path.join(root_dir, 'map', 'user', 'hover', 'country', 'india', 'state')
for state in os.listdir(state_dir):
    state_path = os.path.join(state_dir, state)
    if os.path.isdir(state_path):
        
        for year in os.listdir(state_path):
            year_path = os.path.join(state_path, year)
            if os.path.isdir(year_path):
                
                for json_file in os.listdir(year_path):
                    if json_file.endswith('.json'):
                        with open(os.path.join(year_path, json_file)) as f:
                            data = json.load(f)
                            
                            for district, values in data['data']['hoverData'].items():
                                row_dict = {
                                    'States': state,
                                    'Transaction_Year': year,
                                    'Quarter': int(json_file.split('.')[0]),
                                    'District': district,
                                    'RegisteredUsers': values['registeredUsers'],
                                }
                                data4_list.append(row_dict)

df4 = pd.DataFrame(data4_list)

df4.to_csv('data_map_user.csv', index=False)
```
